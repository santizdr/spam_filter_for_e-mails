```python
contenido = os.listdir(r'C:\Users\santi\Documents\wsp_py\EvalPracticaIA\Enron-Spam\leg√≠timo')
contenido
```

```python
def mail_to_string(path):
    aux = []
    for file in contenido:
        for infile in glob.glob(os.path.join(path, file)):
            file = open(infile, 'r').read()
            aux.append(file)
    return aux
```

```python
ruta = r'C:\Users\santi\Documents\wsp_py\EvalPracticaIA\Enron-Spam\leg√≠timo'
mails = [mail for mail in mail_to_string(ruta)]
```
